```python
import os
import numpy as np
import pandas as pd
from datetime import date


from sklearn.model_selection import KFold, train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import log_loss, roc_auc_score, auc, roc_curve
from sklearn.preprocessing import MinMaxScaler

DATA_ROOT = "C:/Users/User/Documents/GitHub/2nd_ML100Days/data/D48"
```

```python
DATA=os.path.join(DATA_ROOT,'train.csv')
LABELS=os.path.join(DATA_ROOT,'trainLabels.csv')
train_data=pd.read_csv(DATA)
train_labels=pd.read_csv(LABELS)

x_train,x_test,y_train,y_test = train_test_split(train_data,train_labels, test_size = 0.30, random_state = 101)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import datasets, metrics
# 建立模型
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier(min_samples_split=10)
clf3 = RandomForestClassifier(n_estimators=20, max_depth=4)
clf4 = RandomForestClassifier(n_estimators=10, max_depth=4)
clf5 = RandomForestClassifier(n_estimators=20, max_depth=8)
clf6 = GradientBoostingClassifier()

# 訓練模型
clf1.fit(x_train, y_train)
clf2.fit(x_train, y_train)
clf3.fit(x_train, y_train)
clf4.fit(x_train, y_train)
clf5.fit(x_train, y_train)
clf6.fit(x_train, y_train)


# 預測測試集
y_pred1 = clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)
y_pred3 = clf3.predict(x_test)
y_pred4 = clf4.predict(x_test)
y_pred5 = clf5.predict(x_test)
y_pred6 = clf6.predict(x_test)

acc_1= metrics.accuracy_score(y_test, y_pred1)
print("Acuuracy_1: ", acc_1)
acc_2= metrics.accuracy_score(y_test, y_pred2)
print("Acuuracy_2: ", acc_2)
acc_3= metrics.accuracy_score(y_test, y_pred3)
print("Acuuracy_3: ", acc_3)
acc_4= metrics.accuracy_score(y_test, y_pred4)
print("Acuuracy_4: ", acc_4)
acc_5= metrics.accuracy_score(y_test, y_pred5)
print("Acuuracy_5: ", acc_5)
acc_6= metrics.accuracy_score(y_test, y_pred6)
print("Acuuracy_6: ", acc_6)

#print(iris.feature_names)
#print("Feature importance: ", clf.feature_importances_)
#print("Feature importance: ", clf2.feature_importances_)
```

```python
# 設定要訓練的超參數組合
n_estimators = [100, 200, 300]
max_depth = [1, 3, 5]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
grid_search = GridSearchCV(clf6, param_grid, scoring="neg_mean_squared_error", n_jobs=-1, verbose=1)

# 開始搜尋最佳參數
grid_result = grid_search.fit(x_train, y_train)

print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# 使用最佳參數重新建立模型
clf_bestparam = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],
                                           n_estimators=grid_result.best_params_['n_estimators'])

# 訓練模型
clf_bestparam.fit(x_train, y_train)

# 預測測試集
y_pred_bestparam = clf_bestparam.predict(x_test)

acc_pred_bestparam= metrics.accuracy_score(y_test, y_pred_bestparam)
print("Acuuracy_pred_bestparam: ", acc_pred_bestparam)
```

```python
TEST_DATA=os.path.join(DATA_ROOT,'test.csv')
test_data=pd.read_csv(TEST_DATA, header=None)
new_pred = clf_bestparam.predict(test_data)
pred = pd.DataFrame(new_pred)
pred.columns = ['Solution']
pred['Id'] = np.arange(1,len(test_data)+1,1)
pred = pred[['Id', 'Solution']]
pred
```
